```python
missingvalues = ['Unknown/Not Available', 'NaN']

years = [2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]
# print_output = False
# Columns: 'record id', 'report date', 'report time', 'major offense type',
# 'address', 'neighborhood', 'polic precinct', 'police district', 'x coordinate',
# 'y coordinate'
def load_raw_data(print_output, *years):
    df_dict = dict()
    for year in years:
        df_dict[year] = pd.read_csv(os.path.join('crime_incident_data', 'crime_incident_data_' + str(year) + '.csv'))#'/Users/Catalpa/Sites/crime_plot/crime_incident_data/crime_incident_data_' + str(year) + '.csv')
        df_dict[year].columns = [name.lower() for name in df_dict[year].columns]
        df_dict[year]['report date'] = pd.to_datetime(df_dict[year]['report date'])

        if print_output:
            print('Number of dups for', year, ':',  sum(df_dict[year].duplicated()))
            print(df_dict[year][df_dict[year].duplicated()])
            #df_dict[year].drop_duplicates()

            # Show rows where neighborhood is not defined
            print('Number of records without a neighborhood', year, '-')
            print(df_dict[year].loc[df_dict[year]['neighborhood'].isnull()].size)

            print('Number of records total', year, '-')
            print(df_dict[year].size)

            # Filter out null neighborhood rows
            crimes_w_nbhood = df_dict[year][df_dict[year]['neighborhood'].notnull()]
            print('Number of records with neighborhoods', year, '-')
            print(crimes_w_nbhood.size)

        # Drop duplicates
        #df_dict[year].drop_duplicates()
        # Filter out null neighborhood rows
        df_dict[year] = df_dict[year][df_dict[year]['neighborhood'].notnull()]
    return df_dict

crime_years = load_raw_data(True, *years)
```

```python
def neighborhood_merge(dictdf):
    nbhoods = dict()
    union_list = []
    for year, df in dictdf.items():
        #nbhood[year] = df.neighborhood.unique
        # nbhood[year] = pd.unique(df.neighborhood.ravel())
        nbhoods[year] = set(df.neighborhood)
    union_list = set().union(*nbhoods.values())
    master_intersection = union_list
    # Reduce to intersection of all lists:
    for year, s in nbhoods.items():
        master_intersection = master_intersection.intersection(s)
    difference = union_list - master_intersection
    #return (nbhood, union_list)
    return (nbhoods, union_list, master_intersection, difference)

nbhoods, union_list, master_intersection, difference = neighborhood_merge(crime_years)
```

```python
def dictdf_to_neighborhood(dictdf):
    # 102 neighborhoods from master_intersection
    nbhood_list = ['NORTH TABOR', 'PIEDMONT', 'SUNDERLAND', 'MULTNOMAH', 'FOREST PARK', 'OVERLOOK', 'HAYHURST', 'NORTHWEST HTS', 'PARKROSE HGTS', 'HEALY HEIGHTS', 'GRESHAM - CENTENNIAL EAST', 'MONTAVILLA', 'KERNS', 'GLENFAIR', 'KENTON', 'WILKES', 'MADISON SOUTH', 'PEARL', 'GRANT PARK', 'EASTMORELAND', 'RICHMOND', 'HOSFRD-ABRNETHY', 'NORTHWEST', 'HOMESTEAD', 'GRESHAM - WILKES EAST', 'E COLUMBIA', 'ARNOLD CREEK', 'ARGAY', 'SUMNER', 'BROOKLYN', 'ARBOR LODGE', 'SELLWD-MORELAND', 'FAR SOUTHWEST', 'SUNNYSIDE', 'COLLINS VIEW', 'SULLIVANS GULCH', 'CRSTN-KENILWTH', 'ROSE CITY PARK', 'ST JOHNS', 'KING', 'ASHCREEK', 'HOLLYWOOD', 'SOUTH PORTLAND', 'WOODLAND PARK', 'HUMBOLDT', 'LENTS', 'BRIDGETON', 'BOISE', 'SOUTHWEST HILLS', 'MARKHAM', 'ELIOT', 'REED', 'ARDENWALD', 'CENTENNIAL', 'DOWNTOWN', 'ARLINGTON HGHTS', 'BRENTWD-DARLNGT', 'GRESHAM - NORTH CENTRAL', 'CULLY', 'PLEASANT VALLEY', 'SABIN', 'MILL PARK', 'PARKROSE', 'GRESHAM - NORTH GRESHAM', 'VERNON', 'WOODLAWN', 'LAURELHURST', 'HILLSIDE', 'GRESHAM - ROCKWOOD GROUP', 'GOOSE HOLLOW', 'ALAMEDA', 'CONCORDIA', 'BRIDLEMILE', 'CATHEDRAL PARK', 'W PORTLAND PARK', 'RUSSELL', 'CHINA/OLD TOWN', 'BEAUMONT-WILSHR', 'IRVINGTON', 'WOODSTOCK', 'NORTHWEST IND', 'PORTSMOUTH', 'MARSHALL PARK', 'MT TABOR', 'UNIVERSITY PARK', 'LINNTON', 'BUCKMAN-EAST', 'ROSEWAY', 'S BURLINGAME', 'HILLSDALE', 'GRESHAM - DOWNTOWN', 'MAPLEWOOD', 'BUCKMAN-WEST', 'POWELHST-GILBRT', 'MT SCOTT-ARLETA', 'SYLVAN-HIGHLNDS', 'SOUTH TABOR', 'HAYDEN ISLAND', 'LLOYD', 'CRESTWOOD', 'FOSTER-POWELL', 'HAZELWOOD']
    nbhoods_count = pd.DataFrame(index=nbhood_list)
    nbhoods_sum = pd.DataFrame(index=nbhood_list)
    for year, df in dictdf.items():
        df['month'] = df['report date'].apply(lambda d: d.month)
        crimecounts = df[['neighborhood', 'month', 'record id']].groupby(['neighborhood', 'month'], as_index=False)
        df_count = crimecounts.count()
        pivoted = df_count.pivot(index = 'neighborhood', columns='month', values='record id').fillna(0)
        pivoted.columns = [str(year) + '-' + str(month + 1) for month in range(12)]
        nbhoods_sum[year] = pivoted.sum(axis=1)
        pivoted['sum_' + str(year)] = pivoted.sum(axis=1)
        nbhoods_count = pd.concat([nbhoods_count, pivoted.loc[nbhood_list]], axis='neighborhood')
    return (df_count, pivoted, nbhoods_count, nbhoods_sum)
```
